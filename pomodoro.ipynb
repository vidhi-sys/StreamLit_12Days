```python
import streamlit as st
import time
st.header("Pomodoro Focus Timer")
def local_css(file_name):
    with open(file_name) as f:
        st.markdown(f'<style>{f.read()}</style>', unsafe_allow_html=True)

# Create the pomodoro.css file if it doesn't exist to prevent FileNotFoundError
# You can add your CSS content inside the f.write() if you have specific styles.
with open('pomodoro.css', 'w') as f:
    f.write("")
local_css("pomodoro.css")

# Allow user to set duration in minutes
minutes_duration = st.number_input("Enter duration in minutes", min_value=1, value=25)

button_start=st.button("Start")
button_stop=st.button("Stop")
if button_start:
    st.write("Timer started")
    total_seconds = minutes_duration * 60
    for i in range(total_seconds):
        time.sleep(1)
        remaining_seconds_total = total_seconds - i - 1
        remaining_minutes_display = remaining_seconds_total // 60
        remaining_seconds_display = remaining_seconds_total % 60
        st.write(f"{remaining_minutes_display:02d}:{remaining_seconds_display:02d} remaining")
    st.write("Timer finished")
st.audio("https://pixabay.com/sound-effects/search/rain-patter/", format="audio/wav", start_time=0, sample_rate=None, end_time=None, loop=False, autoplay=False)
```
