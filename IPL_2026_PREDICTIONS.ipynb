```python
import streamlit as st
```

```python
import pandas as pd
```

```python
# https://www.kaggle.com/datasets/vinayaks0n1/ipl-2025
# find it on kaggle here
df=pd.read_csv('/content/IPL2025Batters.csv')
```

```python
df.head(5)
```

```python
df = df[['Player Name', 'Runs']]
```

```python
df.head(5)
```

```python
st.write("""
# IPL-2026
Simple Predictor App using StreamLit""")
```

```python
st.line_chart(df, x='Player Name', y='Runs')
```

```python
st.bar_chart(df, x='Player Name', y='Runs')
```

```python
st.scatter_chart(df, x='Player Name', y='Runs',color='#FFC0CB')
```

```python
import plotly.express as px
fig_line = px.line(
    df,
    x='Player Name',
    y='Runs',
    title='Runs by Player (Line Chart)'
)
st.plotly_chart(fig_line, width='stretch')
```

```python
fig_bar = px.bar(
    df,
    x='Player Name',
    y='Runs',
    title='Runs by Player (Bar Chart)'
)
st.plotly_chart(fig_bar, use_container_width=True)
```

```python
fig_scatter = px.scatter(
    df,
    x='Player Name',
    y='Runs',
    title='Runs by Player (Scatter Chart)'
)
st.plotly_chart(fig_scatter, use_container_width=True)
```

```python
fig_line.write_html("line_chart.html")
fig_bar.write_html("bar_chart.html")
fig_scatter.write_html("scatter_chart.html")
```
